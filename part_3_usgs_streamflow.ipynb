```python
import dataretrieval.nwis as nwis
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt

plt.style.use('ggplot')
```

```python
# specify the USGS site code for which we want data.
sites = {'Quinault River':'12039500','Elwa River':'12045500','Big Quilcene River':'12052210','Hoh River':'12041200','Queets River':'12040500','Duckabush River':'12054000'}
start_date = '2015-11-12'
end_date='2015-11-19'

df_discharge = nwis.get_record(sites=list(sites.values()), service='iv', start=start_date, end=end_date, parameterCd='00060')
df_heights = nwis.get_record(sites=list(sites.values()), service='iv', start=start_date, end=end_date, parameterCd='00065')

# # get water quality samples (qwdata)
df2 = nwis.get_stats(sites=list(sites.values())[0], statTypeCd='all')

# # get basic info about the site
df3 = nwis.get_info(sites=list(sites.values()))
```

```python
for k,v in sites.items():
    fig, axs = plt.subplots(2, tight_layout=True, sharex=True, figsize=(10,7))
    da_mi2 = df3[0][df3[0]['site_no']==v]['drain_area_va'].values[0]
    df_discharge.loc[v]['00060'].plot(ax=axs[0], color='purple')

    axs[0].axvline(dt.datetime(2015,11,17,0), color='blue', ls='--')
    axs[0].set_xlabel('Date')
    axs[0].set_ylabel('Discharge (cfs)')
    axs[0].set_title(f'Discharge at {k}: site no. {v}\nDrainage Area = {da_mi2} mi$^2$')
    

    df_heights.loc[v]['00065'].plot(ax=axs[1], color='black')
    
    axs[1].axvline(dt.datetime(2015,11,17,0), color='blue', ls='--')
    axs[1].set_xlabel('Date')
    axs[1].set_ylabel('Gage Height (ft)')
    axs[1].set_title(f'Gage Height at {k}: site no. {v}')
    fig.savefig(f'./figures/{k}.png')
```

```python
df4 = nwis.get_discharge_peaks(list(sites.values()))
df5 = nwis.get_discharge_peaks(list(sites.values()), start_date='2015-11-17', end_date='2015-11-18')
```

```python
for k, v in sites.items():
    fig, ax = plt.subplots(figsize=(5,5))
    ax.scatter(df4[0].loc[v][df4[0].loc[v].index.date != dt.date(2015,11,17)].index, df4[0].loc[v][df4[0].loc[v].index.date != dt.date(2015,11,17)]['peak_va'], edgecolor='blue', facecolor='white')
    ax.scatter(df4[0].loc[v][df4[0].loc[v].index.date == dt.date(2015,11,17)].index, df4[0].loc[v][df4[0].loc[v].index.date == dt.date(2015,11,17)]['peak_va'], c='red', s=100, marker='*', label='2015-11-17 Peak')

    ax.set_xlabel('Year')
    ax.set_ylabel('Peak Discharge ')
    ax.set_title(f'Annual Peak Discharge at {k}')

    fig.savefig(f'./figures/peak_flows/{k}_peaks.png')
```

```python
for k, v in sites.items():
    rating = nwis.get_ratings(site=v)[0]
    fig, ax = plt.subplots(figsize=(5,5))
    idx = df_discharge.loc[v]['00060'].idxmax()
    rating.plot(x='INDEP',y='DEP', ax=ax)
    ax.scatter(df_heights.loc[v]['00065'].loc[idx], df_discharge.loc[v]['00060'].loc[idx], c='red',s=100, marker='*')
    ax.set_xlabel('Gage Height (ft)')
    ax.set_ylabel('Discharge (cfs)')
    ax.set_title(f'Rating curve for USGS Site No. {v}\n{k}')
```
